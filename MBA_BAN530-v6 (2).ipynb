```python
List_Item = []
for i in range(rw):
    temp = []
    for j in item_list:
        if trans_df.loc[i, j] >= 1:
            temp.append(j)
    List_Item.append(temp)

List_Item
```

```python
basket_df = trans_df1[['Item_List']]
basket_df.head()
```

```python
trans_df2= trans_df1.drop(['Item_List'],axis = 1)
trans_df2.head()
```

```python
def encode_units(x):
        if x >= 1:
            return 1
        else: 
            return 0

basket_trans = trans_df2.applymap(encode_units)
basket_trans.head()
```

```python
# Keep rows (transactions) that contain any of the top 5 cheesecakes
cheesecakes_df = basket_trans.loc[basket_trans[cheesecakes].any(axis=1)]
```

```python
#from mlxtend.frequent_patterns import apriori, association_rules

# Run apriori
cheesecakes_frequent_itemsets = apriori(cheesecakes_df, min_support=0.001, use_colnames=True)
cheesecakes_frequent_itemsets
```

```python
# Generate rules
cheesecakes_rules = association_rules(cheesecakes_frequent_itemsets, metric="lift", min_threshold=1)


# Sort rules by lift and support
cheesecakes_rules_sorted = cheesecakes_rules.sort_values(['lift', 'support'], ascending=False).reset_index(drop=True)
```

```python
# Get rules where the antecedent contains any of the top cheesecake names
cheesecakes_rules = cheesecakes_rules_sorted[
    cheesecakes_rules_sorted['antecedents'].apply(lambda x: any(item in cheesecakes for item in x))
]
```
